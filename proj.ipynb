```python
import pandas as pd
import numpy as np

data = pd.read_csv('anime.csv')
data = data[data.type=='TV']
data = data[data.episodes!='Unknown']
data = data.dropna(axis=0)
data.loc[data.episodes.astype(int) <27, 'Length'] = 'short'
data.loc[(data.episodes.astype(int) <100) & (data.episodes.astype(int) >=27) , 'Length'] = 'normal'
data.loc[data.episodes.astype(int) >= 100, 'Length'] = 'long'
data.to_csv('newAnime.csv', sep=',')
```

```python
import pandas as pd

import got3 as got
lst = list()

data = pd.read_csv('projectWithoutSemicolon.csv')

for i, name in enumerate(data.name):
    cnt = 0
    try:
        tweetCriteria = got.manager.TweetCriteria().setQuerySearch(name).setMaxTweets(10000)
        tweet = got.manager.TweetManager.getTweets(tweetCriteria)
        for v in tweet:
            if(name in v.text):
                cnt+=1
        data.at[i,'TweetCnt'] = cnt  
        lst.append(cnt)
    except:
        data.at[i,'TweetCnt'] = cnt
    data.to_csv('projectDataWithTweet.csv', sep=',')
```

```python
/**
1. take user input (separated by commas. Favorite anime title, length of anime that the user prefers, genres that the user prefers)
2. calculate the similarity between the recomended anime and user's favorite anime using Gaussian Radial Basis model
3. old rating, new rating, episodes, and genres are the features used to calculate the grb values. gamma value of 0.1 is used
4. print out the top 10 anime that the user might like based on their preference and favorite anime using GRB model
5. use genre and episodes columns for anime recommendation
6. genre and episodes columns are converted into binary values using minMax Scalar (there are columns for every genre and the code
                                                                                    checks if that anime has certain genre or not)
7. extract the minmax scalar result for the anime that user provided
8. use k nearest neighbor and get the 11 nearest anime (because the first one is the anime that user provided, so we will ignore it
                                                       when we display the top 10 anime to recommend to the user)
**/
```

```python
from sklearn.metrics import pairwise
import pandas as pd
import numpy as np
import math
from operator import itemgetter
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

#can be any input using the anime title in the list above
user_input = "Tonari no Kaibutsu-kun,short,Romance,Slice of Life"

user_input_lst = user_input.split(",")
user_fav_show = user_input_lst[0]
user_prefer_length = user_input_lst[1]
user_fav_genre = user_input_lst[2:]

data = pd.read_csv('projectDataWithNewRating.csv')

fav_show_rating = 0
fav_show_tweets = 0
fav_show_episodes = 0
fav_show_rating = 1

for i, name in enumerate(data.name):
    if(name == user_fav_show):
        fav_show_rating = data.rating[i]
        fav_show_tweetRating = data.NewRating[i]
        fav_show_episodes = data.episodes[i]
        
lst_all = list()
for i, length in enumerate(data.Length):
    if(length == user_prefer_length and data.name[i] != user_fav_show):
        rbf_list = list()
        original_genre_num = len(data.genre[i].split(","))
        match_cnt = 0
        for genre in user_fav_genre:
            if genre in data.genre[i]:
                match_cnt+=1
        new_genere_ratio = match_cnt/original_genre_num  
        new_episodes = data.episodes[i]
        new_tweetRating = data.NewRating[i]
        new_rating = data.rating[i]
        cmp_list = list()
        rating_diff = fav_show_rating - new_rating
        tweetRating_diff = fav_show_tweetRating - new_tweetRating
        episodes_diff = 1 - new_episodes/fav_show_episodes
        genre_diff = 1- new_genere_ratio
        gamma = 0.1
        rate_pow = pow(abs(rating_diff),2)
        tweetRate_pow = pow(abs(tweetRating_diff),2)
        episodes_pow = pow(abs(episodes_diff),2)
        genre_pow = pow(abs(genre_diff), 2)
        rbf = math.exp(-gamma*(rate_pow+tweetRate_pow+episodes_pow+genre_pow))
        rbf_list.append(data.name[i])
        rbf_list.append(rbf)
        lst_all.append(rbf_list)

lst_all = sorted(lst_all, key=itemgetter(1),reverse=True)
temp = ", ".join(user_fav_genre)
print("Using Gaussian Radial Basis Model: ")
print("For someone who likes", user_fav_show, "and wants to watch", user_prefer_length, "anime with these genres:", temp)
print()

for i, rec in enumerate(lst_all):
    if i == 10:
        break
    else:
        print("#",i+1," recommendation:",rec[0])
        print("Similarity rate of the show above and your favorite show:", rec[1])
        print()

anime_features = pd.concat([data["genre"].str.get_dummies(sep=","),data["episodes"]],axis=1)
min_max_scaler = MinMaxScaler()
anime_features = min_max_scaler.fit_transform(anime_features)

index = 0
for i, name in enumerate(data.name):
    if(name == user_fav_show):
        index = i

user_fav_anime = [];
for i, item in enumerate(anime_features):
    if(i == index):
        user_fav_anime = item

neigh = NearestNeighbors(n_neighbors=11)
neigh.fit(anime_features) 
NearestNeighbors(algorithm='auto')
distance, index = neigh.kneighbors([user_fav_anime])

print("------------------------------------------------------------")
print()
print("Using K Nearest Neighbor: ")
print("For someone who likes", user_fav_show)
print()

for rank, i in enumerate(index[0]):
    for j, name in enumerate(data.name):
        if (j == i and rank != 0):
            print("#",rank," recommendation:", name)
            print()
        
```

```python
from sklearn.metrics import pairwise
import pandas as pd
import numpy as np
import math
from operator import itemgetter
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler


f = open("input_sample.txt", "r")

result_file = open("result.txt", "w")

for line in f:
    user_input = line.strip();

    user_input_lst = user_input.split(",")
    user_fav_show = user_input_lst[0]
    user_prefer_length = user_input_lst[1]
    user_fav_genre = user_input_lst[2:]

    data = pd.read_csv('projectDataWithNewRating.csv')

    fav_show_rating = 0
    fav_show_tweets = 0
    fav_show_episodes = 0
    fav_show_rating = 1

    for i, name in enumerate(data.name):
        if(name == user_fav_show):
            fav_show_rating = data.rating[i]
            fav_show_tweetRating = data.NewRating[i]
            fav_show_episodes = data.episodes[i]

    lst_all = list()
    for i, length in enumerate(data.Length):
        if(length == user_prefer_length and data.name[i] != user_fav_show):
            rbf_list = list()
            original_genre_num = len(data.genre[i].split(","))
            match_cnt = 0
            for genre in user_fav_genre:
                if genre in data.genre[i]:
                    match_cnt+=1
            new_genere_ratio = match_cnt/original_genre_num  
            new_episodes = data.episodes[i]
            new_tweetRating = data.NewRating[i]
            new_rating = data.rating[i]
            cmp_list = list()
            rating_diff = fav_show_rating - new_rating
            tweetRating_diff = fav_show_tweetRating - new_tweetRating
            episodes_diff = 1 - new_episodes/fav_show_episodes
            genre_diff = 1- new_genere_ratio
            gamma = 0.1
            rate_pow = pow(abs(rating_diff),2)
            tweetRate_pow = pow(abs(tweetRating_diff),2)
            episodes_pow = pow(abs(episodes_diff),2)
            genre_pow = pow(abs(genre_diff), 2)
            rbf = math.exp(-gamma*(rate_pow+tweetRate_pow+episodes_pow+genre_pow))
            rbf_list.append(data.name[i])
            rbf_list.append(rbf)
            lst_all.append(rbf_list)

    lst_all = sorted(lst_all, key=itemgetter(1),reverse=True)
    temp = ", ".join(user_fav_genre)
    header = "For someone who likes "+user_fav_show+" and wants to watch "+user_prefer_length+ " anime with these genres:" + temp+'\n'
    result_file.write("************************************************\n")
    result_file.write("Using Gaussian Radial Basis Model:\n")
    result_file.write(header)
    result_file.write('\n')

    for i, rec in enumerate(lst_all):
        if i == 10:
            break
        else:
            rate_str = str(rec[1])
            rec = "#"+str(i+1)+" recommendation: "+rec[0]+'\n'
            show = "Similarity rate of the show above and your favorite show:"+rate_str +'\n'
            result_file.write(rec)
            result_file.write(show)
            result_file.write('\n')
            
    anime_features = pd.concat([data["genre"].str.get_dummies(sep=","),data["episodes"]],axis=1)
    min_max_scaler = MinMaxScaler()
    anime_features = min_max_scaler.fit_transform(anime_features)
    
    index = 0
    for i, name in enumerate(data.name):
        if(name == user_fav_show):
            index = i

    user_fav_anime = [];
    for i, item in enumerate(anime_features):
        if(i == index):
            user_fav_anime = item

    neigh = NearestNeighbors(n_neighbors=11)
    neigh.fit(anime_features) 
    NearestNeighbors(algorithm='auto')
    distance, index = neigh.kneighbors([user_fav_anime])
    
    header2 = "For someone who likes " + user_fav_show+ '\n'

    result_file.write("------------------------------------------------------------")
    result_file.write('\n')
    result_file.write("Using K Nearest Neighbor:\n")
    result_file.write(header2)
    result_file.write('\n')
    
    for rank, i in enumerate(index[0]):
        for j, name in enumerate(data.name):
            if (j == i and rank != 0):
                show_knn = "#" + str(rank) + " recommendation: " + name + '\n'
                result_file.write(show_knn)
                result_file.write('\n')
    result_file.write('')
result_file.close()            
```
